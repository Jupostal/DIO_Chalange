```python
import requests
from docx import Document
import os

minha_chave = '89Z9pBh6NI7zbqCgL0nd3rWoJ58WCrGnJFmxkKxFr3ftqDY9B6B7JQQJ99BKACYeBjFXJ3w3AAAbACOGIHoP'
endereco = 'https://api.cognitive.microsofttranslator.com'
local = 'eastus'
linguagem_destino = 'pt-br'

def traduzir_texto(texto, linguagem_destino):
  path = '/translate'
  constructed_url = endereco + path
  headers = {
      'Ocp-Apim-Subscription-Key': minha_chave,
      'Ocp-Apim-Subscription-Region': local,
      'Content-type': 'application/json',
      'X-ClientTraceId': str(os.urandom(16))
  }
  body = [{
      'text': texto
  }]
  params = {
      'api-version': '3.0',
      'from': 'en',
      'to': linguagem_destino
  }
  request = requests.post(constructed_url, params=params, headers=headers, json=body)
  response = request.json()
  return response[0]["translations"][0]["text"]
```

```python
traduzir_texto ("There's only one thing you should know", linguagem_destino )
```

```python
def traduzir_documento(path):
  document = Document(path)
  documento_completo = []
  for paragraph in document.paragraphs:
    texto_traduzido = traduzir_texto(paragraph.text, linguagem_destino)
    documento_completo.append(texto_traduzido)

  traduzido_doc = Document()
  for line in documento_completo:
    print(line)
    traduzido_doc.add_paragraph(line)
  path_traduzido = path.replace(".docx", f'_{linguagem_destino}.docx')
  traduzido_doc.save(path_traduzido)
  return path_traduzido

input_file = "/content/sample_data/Musica.docx"
traduzir_documento(input_file)
```
